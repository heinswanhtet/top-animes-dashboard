# Data Collection

### Constructing Jikan Class Instance

After importing the necessary packages and modules, start creating a Jikan class instance from the jikanpy library. The methods of the created class instance will handle API requests.

```python
# Creating Jikan class instance

jikan = Jikan()
```

### Building Functions

First function: `get_seasonal_animes(year, season)`
- The function requires two parameters: `year` and `season`.
    - `year` parameter takes which year of the data to be searched.
    - `season` parameter takes which season of the data to be searched.
- The function collects all animes data according to its parameters through different pages.
- The function returns all animes based on the input parameters.

```python
def get_seasonal_animes(year, 
                        season):
    '''
    This function will return the seasonal animes list according to its year and season parameter input.
    '''
    
    seasonal_animes = []
    n = 0
    has_next_page = True

    # add to the seasonal_animes list until there is a next page in the pagination
    while has_next_page:
        n += 1
        animes_in_current_page = jikan.seasons(year=year, season=season, page=n)
        seasonal_animes.extend(animes_in_current_page['data'])
        has_next_page = True if animes_in_current_page['pagination']['has_next_page'] else False

        # the function needs to wait about 2 seconds in order not to pass API request limit per second.
        time.sleep(2)

    return seasonal_animes
```

Second function: `get_animes_range_of_years(start_year, end_year = None):`
- The function requires two parameters: `start_year` and `end_year`.
    - `start` parameter takes which lower limit of the year of the data to be searched.
    - `end_year` parameter takes which upper limit of the year of the data to be searched. If it is not provided, make it `None` as default.
- The function collects all animes data in the range of years according to its parameters.
- The function returns all animes based on the input parameters.

```python
def get_animes_range_of_years(start_year, 
                              end_year = None):
    '''
    This function will return all animes list according to its range of years from the parameter input.
    '''

    # checking the end_year parameter is provided or not.
    # If it is not provided, the function assumes the user want to get the data from only one year, so makes start year and end year equal.
    # It it is provided, noting needs to be done and keep the end_year as it is from the input.
    end_year = start_year if not end_year else end_year
    all_animes = []

    # getting data from each year in the input ranges of years
    for year in range(start_year, end_year + 1):
        # In a whole year, there are four season. Therefore, get_seasonal_animes function will help to get data from each season.
        whole_year_animes = get_seasonal_animes(year, 'spring') + \
                            get_seasonal_animes(year, 'summer') + \
                            get_seasonal_animes(year, 'fall') + \
                            get_seasonal_animes(year, 'winter')
        
        all_animes.extend(whole_year_animes)
        print(f'## Collecting {year} data finished!')

    return all_animes
```

### Calling functions

For this project, all animes data in from `2006` to `2022` are needed. 

Therefore, `2006` is set as `start_year` argument and `2022` as `end_year` argument and call the function `get_animes_range_of_years`.

It takes a while to get all necessary data as the required data size is large.


### Building DataFrame

After collecting the data from the API has finished, the data is made as DataFrame with the help of `pandas` library.

```python
anime_data_2006_2022 = get_animes_range_of_years(2006, 2022)
df = pd.json_normalize(anime_data_2006_2022)
```
